```python
# 1. How would you merge the transformed datasets of confirmed cases, deaths, and recoveries on the 'Country/Region' 
# and 'Date' columns to create a comprehensive view of the pandemic's impact?
```

```python
df_c_melt = pd.melt(df_confirmed, id_vars = ["Province/State", "Country/Region", "Lat", "Long"], 
                    var_name = "date", value_name = "cases")

df_c_melt["date"] = pd.to_datetime(df_c_melt["date"])

df_c_melt.isnull().sum()

df_c_melt["Province/State"] = df_c_melt["Province/State"].fillna("All Provinces")

df_c_melt[["Lat", "Long"]] = df_c_melt[["Lat", "Long"]].fillna(-1)


df_c_melt = df_c_melt.sort_values(["Province/State", "Country/Region", "date"])
df_c_melt["cases"] = df_c_melt.groupby(["Province/State", "Country/Region"])["cases"].ffill()
df_c_melt.isnull().sum()


df_c_melt["confirmed_daily_cases"] = df_c_melt.groupby(["Province/State", "Country/Region"])["cases"].diff()
```

```python
df_r_melt = pd.melt(df_recovered, id_vars = ["Province/State", "Country/Region", "Lat", "Long"], 
                    var_name = "date", value_name = "cases")

df_r_melt["date"] = pd.to_datetime(df_r_melt["date"])

df_r_melt.isnull().sum()

df_r_melt["Province/State"] = df_r_melt["Province/State"].fillna("All Provinces")

df_r_melt[["Lat", "Long"]] = df_r_melt[["Lat", "Long"]].fillna(-1)


df_r_melt = df_r_melt.sort_values(["Province/State", "Country/Region", "date"])
df_r_melt["cases"] = df_r_melt.groupby(["Province/State", "Country/Region"])["cases"].ffill()
df_r_melt.isnull().sum()


df_r_melt["recovered_daily_cases"] = df_r_melt.groupby(["Province/State", "Country/Region"])["cases"].diff()
```

```python
df_d_melt = pd.melt(df_deaths, id_vars = ["Province/State", "Country/Region", "Lat", "Long"], 
                    var_name = "date", value_name = "cases")

df_d_melt["date"] = pd.to_datetime(df_d_melt["date"])

df_d_melt.isnull().sum()

df_d_melt["Province/State"] = df_d_melt["Province/State"].fillna("All Provinces")

df_d_melt[["Lat", "Long"]] = df_d_melt[["Lat", "Long"]].fillna(-1)


df_d_melt = df_d_melt.sort_values(["Province/State", "Country/Region", "date"])
df_d_melt["cases"] = df_d_melt.groupby(["Province/State", "Country/Region"])["cases"].ffill()
df_d_melt.isnull().sum()


df_d_melt["deaths_daily_cases"] = df_d_melt.groupby(["Province/State", "Country/Region"])["cases"].diff()
```

```python
df_c_melt = df_c_melt.rename({"cases":"confirmed_cases"}, axis = 1)
df_r_melt = df_r_melt.rename({"cases":"recovered_cases"}, axis = 1)
df_d_melt = df_d_melt.rename({"cases":"deaths_cases"}, axis = 1)


df_c_melt.head()
df_r_melt.head()
df_d_melt.head()
```

```python
# 2. Analyze the monthly sum of confirmed cases, deaths, and recoveries for countries to understand the progression 
# of the pandemic. [From the merged dataset]

df_combined = df_c_melt.merge(df_d_melt[["Province/State", "Country/Region","date","deaths_cases", "deaths_daily_cases"]], 
               on = ["Province/State", "Country/Region", "date"], how = "left")
df_combined
```

```python
# merge combined with recovered dataset
df_merged = df_combined.merge(df_r_melt[["Province/State", "Country/Region","date","recovered_cases", "recovered_daily_cases"]], 
                 on = ["Province/State", "Country/Region", "date"], how = "left")
df_merged
```

```python
# Plot daily recovery rate for all provinces of US
df_us = df_combined[df_merged["Country/Region"] == "US"]
df_us["daily_recovery_rate"] = df_us["recovered_daily_cases"]/df_us["confirmed_daily_cases"]
df_us["daily_recovery_rate"] = df_us["daily_recovery_rate"].fillna(0) 
df_us


plt.figure(figsize = (16,10))
for province in df_us["Province/State"].unique():
    df_sub = df_us[df_us["Province/State"] == province]
    plt.plot(df_sub["date"], df_sub["daily_recovery_rate"])
```

```python
# Plot daily recovery rates of US, France, Italy and Germany - Country level
df_countries = df_combined.loc[df_merged["Country/Region"].isin(["US", "France","Italy", "Germany"])]
df_countries = df_countries.groupby(["Country/Region", "date"])[["recovered_daily_cases", "confirmed_daily_cases"]].sum().reset_index()
df_countries["daily_recovery_rate"] = (df_countries["recovered_daily_cases"]/df_countries["confirmed_daily_cases"]).fillna(0)
df_countries


plt.figure(figsize = (10,5))
for country in ["US", "France", "Italy", "Germany"]:
    df_sub = df_countries[df_countries["Country/Region"] == country]
    plt.plot(df_sub["date"], df_sub["daily_recovery_rate"])
plt.legend(["covid_19_confirmed_v1.csvUS", "France", "Italy", "Germany"])
```

```python
# For ['US', 'France', 'Italy', 'Germany'] find monthly confirmed cases and then plot them
df_countries = df_combined.loc[df_combined["Country/Region"].isin(["US", "France", "Italy", "Germany"])]
df_countries["month"] = df_countries["date"].dt.strftime("%Y-%m")
df_countries = df_countries.groupby(["Country/Region", "month"])[["confirmed_daily_cases"]].sum().reset_index()
df_countries

plt.figure(figsize = (16,9))
for country in ["US", "France", "Italy", "Germany"]:
    df_sub = df_countries.loc[df_countries["Country/Region"] == country]
    plt.plot(df_sub["month"], df_sub["confirmed_daily_cases"])
plt.legend(["US", "France", "Italy", "Germany"])
```
